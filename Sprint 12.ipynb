 Proyecto: Predicción de Recuperación de Oro para Zyfra

## Introducción

En el contexto de la industria pesada, la eficiencia en la extracción de metales preciosos es crucial para optimizar la producción y mejorar la rentabilidad. Zyfra, una empresa especializada en desarrollar soluciones tecnológicas para este sector, busca innovar en sus procesos de extracción y purificación del oro.

El proceso de extracción del oro incluye varias etapas críticas: desde la obtención del mineral y su tratamiento inicial mediante flotación, hasta las fases de purificación que generan concentrados y colas. Cada una de estas etapas afecta la cantidad y la calidad del oro recuperado, por lo que es fundamental contar con herramientas predictivas que permitan optimizar la producción y eliminar parámetros no rentables.

## Objetivo del Proyecto

El objetivo principal de este proyecto es desarrollar un prototipo de modelo de machine learning que prediga la cantidad de oro extraído en dos fases clave:

- **Recuperación del concentrado rougher:** Resultado de la etapa de flotación.
- **Recuperación final:** Resultado del proceso de purificación.

Estas predicciones se utilizarán para:

- Optimizar el proceso de extracción.
- Detectar y descartar parámetros que afecten negativamente la eficiencia.
- Proporcionar una herramienta de apoyo a la toma de decisiones en la industria.

## Estrategias y Metodología

Para alcanzar este objetivo, se seguirá un enfoque estructurado que abarca las siguientes fases:

1. **Preparación y Exploración de Datos:**
   - **Carga de datos:** Se trabajará con tres conjuntos de datos (entrenamiento, prueba y completo), en los cuales se registran diversas variables obtenidas durante el proceso de extracción y purificación.
   - **Análisis exploratorio:** Se examinarán las variables, se evaluará la calidad de los datos y se identificarán posibles inconsistencias, valores nulos o outliers que puedan afectar el desempeño del modelo.

2. **Preprocesamiento:**
   - **Imputación y limpieza:** Se aplicarán técnicas de imputación para tratar los valores faltantes y se eliminarán o transformarán aquellos datos que se consideren anómalos.
   - **Selección y transformación de características:** Se identificarán las variables más relevantes para la predicción y se realizarán transformaciones necesarias para mejorar la relación entre las features y los objetivos.

3. **Modelado:**
   - **Construcción del modelo:** Se abordará la tarea como un problema de regresión multisalida, utilizando diferentes algoritmos de machine learning (por ejemplo, Regresión Lineal y Random Forest).
   - **Optimización de hiperparámetros:** Se emplearán técnicas de búsqueda (como GridSearchCV) para ajustar los parámetros del modelo y maximizar su desempeño.
   - **Evaluación del modelo:** Se utilizará la métrica sMAPE (error medio absoluto porcentual simétrico) para evaluar la precisión del modelo, tanto en validación cruzada como en un conjunto de validación independiente.

4. **Predicción y Resultados:**
   - **Aplicación en datos de prueba:** Una vez validado el modelo, se utilizará para generar predicciones en el conjunto de datos de prueba, que no incluye las variables objetivo.
   - **Interpretación de resultados:** Se analizarán las predicciones obtenidas y se compararán con los valores esperados, proporcionando una herramienta que permita identificar oportunidades de mejora en el proceso de extracción.

Lo que voy a realizar a continuación es comprobar que el cálculo de la recuperación sea el correcto, para esto voy a generar una nueva columna en el dataframe de entrenamiento y comparar los valores calculados con los valores de la columna rougher.output.recovery para determinar que sean los adecuados. 

```python
F = "rougher.input.feed_au"
C = "rougher.output.concentrate_au"
T = "rougher.output.tail_au"
R = "rougher.output.recovery"

# Calcular la recuperación usando la fórmula: C(F-T)/(F(C-T))
# Se evita la división por cero (esto se puede manejar mediante un filtro si es necesario)
train_df["calc_recovery"] = 100*(train_df[C]* (
    train_df[F] - train_df[T])) /(train_df[F]* (train_df[C] - train_df[T]))


# Filtrar las filas con datos completos en ambos valores
mask = train_df[R].notna() & train_df["calc_recovery"].notna()

# Calcular el MAE entre la recuperación registrada y la calculada
mae = mean_absolute_error(
    train_df.loc[mask, R], train_df.loc[mask, "calc_recovery"])
print("MAE entre la recuperación registrada y la calculada:", mae)

# Visualizar la comparación mediante un scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(train_df.loc[mask, R],
            train_df.loc[mask, "calc_recovery"], alpha=0.5)
plt.xlabel("Recuperación Registrada")
plt.ylabel("Recuperación Calculada")
plt.title("Comparación: Recuperación Registrada vs Calculada")
# Dibujar la línea de identidad
plt.plot([0, 100], [0, 100], 'r--')
plt.show()
```

```python

# Definimos para cada metal las 8 columnas (etapas) correspondientes:
metal_columns = {
    "Au": [
        "rougher.input.feed_au",
        "rougher.output.tail_au",
        "rougher.output.concentrate_au",
        "primary_cleaner.output.tail_au",
        "primary_cleaner.output.concentrate_au",
        "secondary_cleaner.output.tail_au",
        "final.output.tail_au",
        "final.output.concentrate_au"
    ],
    "Ag": [
        "rougher.input.feed_ag",
        "rougher.output.tail_ag",
        "rougher.output.concentrate_ag",
        "primary_cleaner.output.tail_ag",
        "primary_cleaner.output.concentrate_ag",
        "secondary_cleaner.output.tail_ag",
        "final.output.tail_ag",
        "final.output.concentrate_ag"
    ],
    "Pb": [
        "rougher.input.feed_pb",
        "rougher.output.tail_pb",
        "rougher.output.concentrate_pb",
        "primary_cleaner.output.tail_pb",
        "primary_cleaner.output.concentrate_pb",
        "secondary_cleaner.output.tail_pb",
        "final.output.tail_pb",
        "final.output.concentrate_pb"
    ]
}


def melt_metal_columns(df, cols):
    """
    Recibe el DataFrame y la lista de columnas de un metal.
    Devuelve un DataFrame en formato largo con columnas: 'Stage' y 'Concentration'.
    """
    data = []
    for col in cols:
        if col in df.columns:  # Verificar que la columna exista en df
            stage_name = col
            for val in df[col]:
                data.append({"Stage": stage_name, "Concentration": val})
    return pd.DataFrame(data)


# Generar gráficas separadas, una para cada metal
for metal, cols in metal_columns.items():
    # Convertir las columnas del metal a formato largo
    df_melted = melt_metal_columns(train_imputed, cols)

    # Crear una figura y eje independiente para cada metal
    plt.figure(figsize=(10, 6))
    sns.boxplot(x="Stage", y="Concentration", data=df_melted)
    plt.title(f"Distribución de {metal} en las 8 columnas (etapas)")
    plt.xlabel("Etapa/Columna")
    plt.ylabel("Concentración")
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()
```

Ahora podemos proceder al entrenamiento del modelo para lo cual primero escribimos la función para calcular sMAPE que es la siguiente: 

```python

def sMAPE(y_true, y_pred):
    """Calcula el sMAPE estándar entre y_true y y_pred."""
    epsilon = 1e-8  # para evitar división por cero
    return 100 * np.mean(2 * np.abs(y_true - y_pred) / (np.abs(y_true) + np.abs(y_pred) + epsilon))


def sMAPE_final(y_true_rougher, y_pred_rougher, y_true_final, y_pred_final):
    """
    Calcula el sMAPE final ponderado:
    sMAPE_final = 0.25 * sMAPE(rougher) + 0.75 * sMAPE(final)
    """
    smape_rougher = sMAPE(y_true_rougher, y_pred_rougher)
    smape_final_val = sMAPE(y_true_final, y_pred_final)
    return 0.25 * smape_rougher + 0.75 * smape_final_val
```
